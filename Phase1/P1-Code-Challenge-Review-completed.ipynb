```python
first_20_songs[0]["artists"]
```

```python
# Your code here:
song_dictionary = {}
for song in first_20_songs:
    artist_list = []
    for artist in song['artists']:
        artist_list.append(artist['name'])
    song_dictionary[song['name']] = tuple(artist_list)
song_dictionary
```
